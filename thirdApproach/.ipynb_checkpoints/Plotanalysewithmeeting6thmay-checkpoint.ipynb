```python
import pandas as pd
from matplotlib import pyplot  as plt
import numpy as np
```

```python
a=pd.read_csv("CITandourMix.csv")
```

```python
categories=np.array(a["GroundTruth"])
colormap=np.array(['y','b'])
plt.scatter(a["p_cit"],a["pour"],c=colormap[categories])
plt.title("p values compare")
plt.xlabel("CIT")
plt.ylabel("Our")
#plt.legend(loc='ll', label='legend labe')
plt.colorbar()
```
