```python
########## making table of non linearity using ks test ##########
genotype=pd.read_csv("genotypes_binary_strongest_eqtl.csv",header=0,index_col=0)
expression_data=pd.read_csv("expression_reordered_01.csv",index_col=0,header=0)
causal_conf=read_configuration("./yeastgt_1_wilko1234_ready.txt")
indp_conf=read_configuration("./yeastgt_0_wilko1234_ready.txt")
```

```python
mix=pd.concat([causal_conf,indp_conf],axis=0)
```

```python
mix
```

```python
mix_group = mix.groupby(0)
```

```python
set_of_values = mix_group.agg(set)
```

```python
set_of_values
```

```python
def set_to_list(row):
    return [list(item) for item in row]

# Apply the function to each row of the DataFrame
set_of_values = set_of_values.apply(set_to_list, axis=1)
```

```python
data_dict={}
# Iterate through each key in the dictionary
for index in set_of_values.index:
    # Convert the set to a list and assign it back to the key in the dictionary
    data_dict[index] = [element for innerList in set_of_values[index] for element in innerList]
```

```python
data_dict
```

```python
combined_dfs=[]
for key, values in data_dict.items():
    # Get the gene expression values corresponding to the current key
    current_df={}
    key_without_index = genotype[key].reset_index(drop=True)
    current_df[key]=key_without_index
    for val in values: 
        val_without_index = expression_data[val].reset_index(drop=True)
        current_df[val]=val_without_index

    df = pd.DataFrame(current_df)
    
    # Append the DataFrame to the list of combined DataFrames
    combined_dfs.append(df)
```

```python
#insertion ordering is present
combined_dfs[1]
```

```python

# Assuming dfs is your list of DataFrames
for df in combined_dfs:
    # Extract the name of the second column
    filename = "./mrpc_yeast/"+ df.columns[1] + ".csv"
    # Remove row names/index before saving
    df.to_csv(filename, index=False)
```

```python
    
for certain_fdr_value in [0,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1]:
    result_dict = {}

    directory = "/data/users/cs20s037/CITNonLinear/conclude/newmethod/simulation_final/results/journal/simulation/MRPC/yeast_results/"

    # List all files in the directory
    files = os.listdir(directory)

    # Filter out CSV files
    csv_files = [file for file in files if file.endswith('.csv')]


    # Read each CSV file using pandas
    for csv_file in csv_files:
        #print(csv_file)
        file_path = os.path.join(directory, csv_file)
        df = pd.read_csv(file_path,header=0)
        # Assuming df is your DataFrame with column 'fdr' and other columns with 'edge names' wording
    # 'certain_fdr_value' is the value you want to filter for in the 'fdr' column

    # Step 1: Filter rows based on certain value of 'fdr'
    
        filtered_df = df[df['fdr'] == certain_fdr_value]
        # Step 2: Ensure there's only one row for a given fdr
        assert len(filtered_df) == 1, "There should be only one row for the given fdr value"
        
        # Step 3: Iterate through the single row of filtered_df and process 'edge names' columns

        row = filtered_df.iloc[0]  # Extracting the single row from filtered_df

        for column in row.index:
            if 'edge names' in column and not pd.isnull(row[column]):
                #print(row[column])
                splits = row[column].split('|')
                #print(splits)
                key=splits[0]
                value = splits[1]
                #print(key)
                if key not in result_dict:
                    result_dict[key] = set()
                result_dict[key].add(value)

    # Function to check presence of values in dictionary
    def check_presence(row):
        if row[1] in result_dict:
            if row[2] in result_dict[row[1]]:
                return 1  # Present
        return 0  # Absent



    results_pos_conf=causal_conf.copy()
    results_neg_conf=indp_conf.copy()
    # Add row number and FDR cutoff
    results_pos_conf.insert(0, 'trio', results_pos_conf.index + 1)  # Assuming 1-based indexing
    results_pos_conf.insert(1, 'fdr', certain_fdr_value)
    results_neg_conf.insert(0, 'trio', results_neg_conf.index + 1)  # Assuming 1-based indexing
    results_neg_conf.insert(1, 'fdr', certain_fdr_value)
    # Add column to mark presence/absence
    #print(results_pos_conf)
    results_pos_conf['Present'] = results_pos_conf.apply(check_presence, axis=1)
    results_neg_conf['Present'] = results_neg_conf.apply(check_presence, axis=1)
    #print(result_dict)
    causal_final=causal_final.append(results_pos_conf)
    indp_final=indp_final.append(results_neg_conf)

```

```python
column_mapping = {0: 'L', 1: 'A',2:'B'}

# Rename the set of columns
causal_final = causal_final.rename(columns=column_mapping)
indp_final = indp_final.rename(columns=column_mapping)
```
